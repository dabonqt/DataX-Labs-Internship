```python
# Loading the dataset
df = pd.read_csv('marketing_campaign.csv', sep='\t')
```

```python
# Show a quick table summary
df.head()
```

```python
# Shows column names, data types, and non-null counts
df.info()
```

```python
# Check and display summary statistics for numeric columns
df.describe()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Check unique values
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")
```

```python
# Check value count
for col in df.columns:
    print(f"--- Value Counts for {col} ---")
    print(df[col].value_counts())
    print("\n")
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
#Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace(r'^(mnt|num|acceptedcmp)(.+)$', r'\1_\2', regex=True)
```

```python
# Check current data types
df.dtypes
```

```python
# Convert column to a number
df['income'] = pd.to_numeric(df['income'], errors='coerce')
# Replace missing values from 'Income' column with a placeholder
df['income'] = df['income'].fillna(df['income'].median())
```

```python
# Check outliers in income
print(df[df['income'] > 200000][['id', 'income']])
```

```python
# Drop any row where income is above the highest limit
df = df[df['income'] <= 200000]
print(f"Remaining rows: {len(df)}")
```

```python
# Convert 'dt_customer' data type
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')
```

```python
# Replace/drop unwanted values in 'Marital Status' column
df['marital_status'] = df['marital_status'].replace('Alone', 'Single')
df = df[~df['marital_status'].isin(['YOLO', 'Absurd'])]
```

```python
# See rows with problematic birth year
print(df[df['year_birth'] < 1930][['id', 'year_birth']])
```

```python
# Drop rows with unrealistic birth year
df = df[df['year_birth'] >= 1930]
# Confirm they're gone
print(f"Remaining rows: {len(df)}")
```

```python
# Drop useless columns
df = df.drop(columns=['z_costcontact', 'z_revenue'])
```
